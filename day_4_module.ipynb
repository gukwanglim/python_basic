```python
url = 'https://www.weather.go.kr/w/weather/forecast/mid-term.do'

resp = requests.get(url)

soup = BeautifulSoup(resp.text)

loca_city = []
loca_wf = []
loca_tmn = []
loca_tmx = []

for data in soup.findAll('table', class_='table-col'):
    for line in data.findAll('tr'):

        name = line.find('td', class_='midterm-city')

        if name:
            loca_city.append(name.get_text().strip('\n'))
    # # print(loca_city)


        weather = line.find('i', class_='wic large NB08')

        if weather:
            loca_wf.append(weather.get_text().strip('\n'))
    # print(loca_wf)


        t_min = line.find('span', class_='tmn')

        if t_min:
            loca_tmn.append(t_min.get_text().strip('\n'))
    # print(loca_tmn)


        t_max = line.find('span', class_='tmx')

        if t_max:
            loca_tmx.append(t_max.get_text().strip('\n'))
    # print(loca_tmx)

for city, weather, t_min, t_max in zip(loca_city, loca_wf, loca_tmn, loca_tmx):
#     print(city, '-', weather, ':', t_min, '~', t_max)
    print()
    print('도시 \t : ', city, '\n' '날씨 \t : ', weather, '\n'
          '최저기온 : ', t_min, '\n' '최고기온 : ', t_max)


# for location in soup.select('location'):
#     print('도시 : ', location.select_one('city').string)
#     print('날씨 : ', location.select_one('wf').string)
#     print('최저기온 : ', location.select_one('tmn').string)
#     print('최고기온 : ', location.select_one('tmx').string)
#     print()
```
