```python
# 문) 키보드에서 숫자 부호 숫자를 입력받아 계산하는 프로그램 작성
# 10 + 20 -> 10 + 20 = 30 형식으로 출력
# 처음 자료에 'q'가 입력되면 프로그램 종료
# 부호는 '+-*/'만 사용, 이외의 부호가 입력되면 계속
# 처음 자료가 숫자 또는 'q'가 아니면 계속 입력
```

```python
# 1. 반복문
# 2. 계산식을 입력 받음, 자료를 list로 변환 split()
# 3. 처음 자료가 'q'이면 반복문 탈출 아니면 처음 자료와 마지막 자료가 숫자인지 확인 -> int로 변환
# 4. 부호에 따라 계산식 출력

while True:
    data = input('계산식 : ').split()
    if not data or data[0] == 'q':
        print('계산식을 종료합니다.')
        break
    elif len(data) < 3:
        print('정확히 입력해주세요.')
        continue
    elif data[1] not in '+-*/':
        print('부호를 다시 입력해주세요.')
        continue
    elif not (data[0].isdigit() and data[2].isdigit()):
        print('숫자를 다시 입력해주세요.')
        continue
    num1 = int(data[0])
    buho = data[1]
    num2 = int(data[2])
    if buho == '+':
        print(num1 + num2)
    elif buho == '-':
        print(num1 - num2)
    elif buho == '*':
        print(num1 * num2)
    else:
        print(num1 / num2)    
        
```

```python
while True:
    input_data = input("계산식 입력 (예: 10 + 20), 종료:'q'").split()
    if input_data[0] == 'q': # 'q'가 입력되면 종료
        break
    if len(input_data) < 3:  # 입력된 데이터가 3개 미만이면 계속 입력
        continue
    # 처음과 마지막 자료가 숫자가 아니면 다시 입력
    if not input_data[0].isdigit() or not input_data[2].isdigit():
        continue
    # 부호가 '+-/'가 아니면 다시 입력
    if not input_data[1] in '+-/':
        continue

    num1 = int(input_data[0])
    buho = input_data[1]
    num2 = int(input_data[2])
    # 계산식 출력
    if buho == '+':
        result = num1 + num2
    elif buho == '-':
        result = num1 - num2
    elif buho == '*':
        result = num1 * num2
    else:
        result = num1 / num2
    print("{} {} {} = {}".format(num1, buho, num2, result))

print("프로그램 종료")
```

```python
# 이름과 성적을 입력받아 딕셔너리에 저장한 후,
# 인원수, 성적 총합계, 평균을 구하세요.
# 입력의 종류는 이름에 'q'가 입력되면 입력 종료.
```

```python
# 빈 딕셔너리 만들기
dict_data = {}

# 이름과 성적 입력 받아 딕셔너리에 저장.
while True:
    input_data = input('이름과 성적을 입력하십시오(예 : 홍길동 55) : ').split()
        # 이름에 'q'가 입력되면 입력 종료.
    if input_data[0] == 'q':
        print('입력을 종료합니다.')    # 순서 중요. 이 부분이 밑으로 내려가면 'q'만을 작성하고
        break                          # 점수 부분이 비어있다며 에러가 뜬다.
    elif len(input_data) < 2:
        continue
    elif not input_data[1].isdigit():
        continue
    name = input_data[0]               # 이 부분이 위로 올라가면 'q' 사용했을 시 오류.
    score = int(input_data[1])
    dict_data[name] = score
    
    # 딕셔너리를 가지고 인원수, 성적 총합계, 평균 구하기
    count = 0
    total = 0
    for key in dict_data:                       # 이 부분이 중요하다.(딕셔너리로 총합, 평균)
        count += 1
        total += dict_data[key]
                                                # 소수점 이하 두 자리만 보여준다.
print("학생 수 : {}명, 총합계 : {},  평균 : {:3.2f}".\
      format(count,total, total/count))
    
```

```python
student = {}
# 자료 입력 : 이름에 'q'가 입력되면 입력 종료
while True:
    input_data = input("이름 성적 입력 (종료는 'q'): ").split()
    if input_data[0] == 'q':
        break
    elif len(input_data) < 2:
        continue
    elif not input_data[1].isdigit():
        continue
#     name = input_data[0]
#     score = int(input_data[1])
#     student[name] = score
    student[input_data[0]] = int(input_data[1])

# 인원수, 성적 총합계, 평균을 구하세요
cnt = 0
total = 0
for key in student:
    cnt += 1
    total += student[key]

# 계산 결과 출력
print("학생 수 : {}명, 총합계 : {},  평균 : {:3.2f}".\
      format(cnt,total, total/cnt))
```

```python
# 문) 1부터 100까지의 홀수의 합과 짝수의 합을 구하시오

numbers = range(100 + 1)
odd = [i for i in numbers if i % 2 == 1]
even = [j for j in numbers if j % 2 == 0]

print('1부터 100까지 홀수의 합 : ', sum(odd))
print('1부터 100까지 짝수의 합 : ', sum(even))
```
