```python
# titanic data load

import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
df.head()
```

```python
# age와 fare 컬럼만 추출

df = df.loc[ : , ['age', 'fare']]
df.head()
```

```python
# ten 컬럼을 추가

df['ten'] = 10
df.head()
```

```python
def add_10(n):
    return n + 10

def add_two_obj(a, b):
    return a + b


print(add_10(20)); print()

print(add_two_obj(10, 10))
```

```python
# 시리즈 객체의 각 원소에 함수를 매핑 : .apply(매핑함수), apply는 시리즈의 함수로 시리즈를 return

sr1 = df['age'].apply(add_10)                # 시리즈를 반환, add_10(n)에서 n의 값에 df['age'] 적용


print(df['age'].head(3)); print()
print('--------')

print(sr1.head(3))
```

```python
# 시리즈 객체와 숫자를 적용한 함수 add_two_obj(a, b)

sr2 = df['age'].apply(add_two_obj, b = 10)   # add_two_obj(a, b)의 a의 값에 df['age'] 적용

print(sr2.head(3))
```

```python
# 매핑함수를 lambda 함수로 정의

sr3 = df['age'].apply(lambda x: add_10(x))  # x <- df['age'] 각각의 원소

print(sr3.head(3))
```

```python
# 데이터프레임의 각 원소에 함수 매핑 : .applymap(매핑함수)

df = sns.load_dataset('titanic').loc[ : , ['age', 'fare']]
df.head(3)
```

```python
df_map = df.applymap(add_10)    # age, fare 각각에 add_10 함수 적용
df_map.head(3)                  # applymap은 데이터프레임을 return
```

```python
# apply 함수를 데이터프레임에 적용 : .apply(매핑함수, axis = 0)      # 계산하는 거니까 0이 세로
                                        # -> 컬럼을 시리즈로
def missing_value(serise):
    return serise.isnull()

result = df.apply(missing_value, axis = 0)
result.head(3)
```

```python
# 최대값 - 최소값

def max_min(x):
    return x.max() - x.min()

result = df.apply(max_min)        # 데이터프레임의 열 단위로 함수 적용
result
```

```python
# 데이터프레임의 행 단위로 원소에 함수를 적용

df['ten'] = 10
df.head(3)
```

```python
df['result'] = df.apply(lambda x: add_two_obj(x['age'], x['ten']), axis = 1)
df.head(3)
```

```python
# fare 컬럼에 10을 더해서 fare_add 컬럼으로 추가

df['fare_add'] = df.apply(lambda x: x['fare'] + x['ten'], axis = 1)
df.head(3)
```

```python
# .pipe() : 데이터프레임의 객체를 함수에 매핑할 때, 뒤에 따라오는 매핑함수의 return 에 따라 변환

df = df.loc[ : , ['age','fare']]
df.head(3)
```

```python
def missing_value(x):
    return x.isnull()

def missing_count(x):
    return missing_value(x).sum()

def total_missing_number(x):
    return missing_count(x).sum()
```

```python
result_df = df.pipe(missing_value)
print(result_df.head(3)); print()

print(type(result_df))
```

```python
result_sr = df.pipe(missing_count)
print(result_sr); print()

print(type(result_sr))
```

```python
result_total = df.pipe(total_missing_number)
print(result_total, type(result_total))
```

```python
# 데이터프레임 만들기

df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['c2', 'c3', 'c4', 'c5'],
                    'd': ['d2', 'd3', 'd4', 'd5']},
                    index=[2, 3, 4, 5])
```
