# 모듈로 홈페이지(url)에 있는 자료 가져오기

```python
# 필요한 모듈 불러내기

import requests
from bs4 import BeautifulSoup
```

```python
# 자료를 뽑아올 홈페이지(기상청 날씨)

url = 'https://www.weather.go.kr/w/weather/forecast/mid-term.do'

# url을 가져오는 requests, BeatifulSoup4 기능

resp = requests.get(url)          # url의 자료를 가져올 것을 요청한다.

soup = BeautifulSoup(resp.text)   # 위의 요청을 허락하면 스프를 사용할거다

# 필요한 정보를 리스트로 만들기 위해 빈 리스트를 만든다.

loca_city = []
loca_wf = []
loca_tmn = []
loca_tmx = []

# 모든 라인에서 'tr'에 관련된 내용을 찾아낼 것이다.

for line in soup.findAll('tr'):
    
    # 홈페이지에 들어가 오른쪽 상단의 점 세 개에서 '도구 더보기' - '개발자 도구'
    # 왼쪽 상단 마우스 모양 이모티콘을 사용하여 내가 얻고 싶은 내용을 클릭
    # 클릭한 내용의 주소와 클래스가 보인다.
    
    # 간단한 이름을 설정한 후, 라인에서 찾기로 주소, 클래스를 넣는다.
    
    name = line.find('td', class_='midterm-city')
    
    # if 구문을 이용하여 내가 찾는 내용들을 모두 처음 만들어 놓은 리스트에 넣는다.
    
    if name:
        loca_city.append(name.get_text().strip('\n'))
# # print(loca_city)

    # 필요한 자료의 수만큼 위의 내용을 반복
    
    weather = line.find('i', class_='wic large NB08')
    
    if weather:
        loca_wf.append(weather.get_text().split('\n'))
# print(loca_wf)


    t_min = line.find('span', class_='tmn')
    
    if t_min:
        loca_tmn.append(t_min.get_text().split('\n'))
# print(loca_tmn)


    t_max = line.find('span', class_='tmx')
    
    if t_max:
        loca_tmx.append(t_max.get_text().split('\n'))
# print(loca_tmx)

    # 모든 자료를 수집했으면 for 구문으로 보기 좋게 정리
    
    # 순위표가 나온다면 
    # for i, value in enumerate(리스트):
    # print(range(i + 1), value)
    #를 활용하여 순서대로 정리
    
    # 자료가 여러개로 나온다면
    # for list_a, list_b in zip(object_a, object_b):
    # print(list_a, ':', list_b)
    # 를 사용하여 정리
    
    # 여러 자료가 순위표로 나온다면
    # for i, (list_a, list_b) in enumerate(zip(object_a, object_b)):
    # print(range(i + 1), list_a, ':', list_b)
    # 를 사용하여 정리

for city, weather, t_min, t_max in zip(loca_city, loca_wf, loca_tmn, loca_tmx):
#     print(city, '-', weather, ':', t_min, '~', t_max)
    print()
    print('도시 : ', city, '\n' '날씨 : ', weather, '\n'
          '최저기온 : ', t_min, '\n' '최고기온 : ', t_max)


# for location in soup.select('location'):
#     print('도시 : ', location.select_one('city').string)
#     print('날씨 : ', location.select_one('wf').string)
#     print('최저기온 : ', location.select_one('tmn').string)
#     print('최고기온 : ', location.select_one('tmx').string)
#     print()
```

```python
# 이 방법이 훨씬 쉽고 간단하다

url = 'https://www.weather.go.kr/w/weather/forecast/mid-term.do'

soup = request.urlopen(url)
data = BeautifulSoup(soup)

# 각 지역의 최저, 최고 기온

for table in data.findAll('table', class_="table-col"):
    for line in table.findAll('tr'):
        if line.find('td', class_='midterm-city'):
            print(line.find('td', class_='midterm-city').get_text())
            print('최저기온 : ', line.find('span', class_='tmn').get_text())
            print('최고기온 : ', line.find('span', class_='tmx').get_text())
```
