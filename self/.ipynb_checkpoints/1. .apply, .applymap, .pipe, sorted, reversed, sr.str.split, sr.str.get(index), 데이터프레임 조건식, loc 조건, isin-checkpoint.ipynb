```python
import seaborn as sns
import pandas as pd
```

```python
# titanic load

ti = sns.load_dataset('titanic')
ti.head()
```

```python
# age와 fare 컬럼만 추출

df = ti.loc[ : , ['age', 'fare']]
df.head()
```

```python
# ten 컬럼을 추가

df['ten'] = 10              # 이렇게 숫자 하나만 넣으면 모든 행에 숫자가 들어감
df.head()
```

```python
# 함수 정의

def add_10(n):
    return n + 10

def add_two_obj(a, b):
    return a + b
```

# 시리즈 객체의 각 원소에 함수를 매핑 : .apply(매핑함수), apply는 시리즈의 함수로 시리즈를 return

```python
sr1 = df['age'].apply(add_10)    # 'age' 컬럼에 add_10 함수를 적용하라
sr1.head()                       # apply가 적용되면 시리즈의 형태로 반환한다.
            
                                 # add_10(n)에서 n의 값에 df['age']를 적용
```

```python
# 시리즈 객체와 숫자를 적용한 함수 add_two_obj(a, b)

sr2 = df['age'].apply(add_two_obj, b = 10)    # 'age' 컬럼에 add_two_obj 함수를 적용하라
sr2.head()                                    
                                              # add_two_obj(a, b)에서 a의 값에 df['age']를 적용
```

```python
# 매핑함수를 lambda 함수로 정의

sr3 = df['age'].apply(lambda x: add_10(x))
sr3.head()                                 # x의 값에 df['age']를 적용
```

# 데이터프레임의 각 원소에 함수 매핑 : .applymap(매핑함수)

```python
df1 = ti.loc[ :, ['age', 'fare']]
df1.head()
```

```python
df1_1 = df1.applymap(add_10)          # 데이터프레임에 apply를 적용하고 싶으면 applymap을 사용
df1_1.head()                          # applymap은 모든 컬럼에 매핑함수를 적용

                                      # 데이터프레임으로 return 되지만 apply의 기본은 Serise이다
```

```python
# apply 함수를 데이터프레임에 적용 : .apply(매핑함수, axis = 0)      # 계산하는 거니까 0이 세로
                                        # -> 컬럼을 시리즈로
    
def missing_value(serise):
    return serise.isnull()

result = df.apply(missing_value, axis = 0)
result.head(3)
```

```python
# 최대값 - 최소값

def max_min(x):
    return x.max() - x.min()

result = df.apply(max_min)        # 데이터프레임의 열 단위로 함수 적용
result                                       # apply는 시리즈에 적용되는 것으므로 간단하게 생각
```

```python
# 데이터프레임의 행 단위로 원소에 함수를 적용

df.head()
```

```python
df['result'] = df.apply(lambda x: add_two_obj(x['age'], x['ten']), axis = 1)
df.head(3)                  
                                # 'age'와 'fare'의 값을 더해 'result'로 만들어라(계산이므로 가로)
```

```python
# fare 컬럼에 10을 더해서 fare_add 컬럼으로 추가

df['fare_add'] = df.apply(lambda x: add_two_obj(x['fare'], b = 10), axis = 1)
df.head()
                            # apply는 시리즈의 명령어로 새로운 컬럼을 만들기 위해서는 axis=1이 필요
```

# .pipe() : 데이터프레임의 객체를 함수에 매핑할 때, 뒤에 따라오는 매핑함수의 return 에 따라 변환

```python
df2 = ti.loc[ : , ['age', 'fare']]
df2.head()
```

```python
def missing_value(x):
    return x.isnull()                     # x에 nan이 존재하면 True로 반환

def missing_count(x):
    return missing_value(x).sum()         # 위 함수에서 찾은 True의 개수를 세어 반환(각 컬럼별)

def total_missing_number(x):
    return missing_count(x).sum()         # 위 함수에서 세어놓은 수를 합침(컬럼 + 컬럼)
```

### 데이터프레임 return

```python
result_df2 = df2.pipe(missing_value)         # pipe 뒤의 함수 isnull은 데이터프레임의 함수이다.
result_df2.head()
```

```python
type(result_df2)                              # 따라서 데이터프레임을 return 한다.
```

### 시리즈 return

```python
result_sr = df2.pipe(missing_count)           # pipe 뒤의 함수 sum은 시리즈의 함수로 사용되었다.   
result_sr.head()
```

```python
type(result_sr)                               # 따라서 시리즈를 return 한다.
```

### numpy return

```python
df2.info()                                    # df2를 확인하면 fare가 age보다 177개 많음
                                              # 즉, age는 177개의 nan을 가지고 있다 할 수 있음
```

```python
result_total = df2.pipe(total_missing_number)       # pipe 뒤의 함수는 numpy 함수이다.
result_total
```

```python
type(result_total)                                  # 따라서 numpy를 return한다.
```

# isin() 메소드 활용

```python
# 데이터프레임 열 객체 : .isin(추출 값의 리스트), ()에는 리스트가 아닌 시리즈가 들어가도 괜찮
# 타이타닉에 함께 탑승한 형제 자매의 수가 3, 4, 5인 승객만 추출(sibsp)

ti.head(10)
```

```python
mask1 = ti['sibsp'] == 3              # 이런 식으로 df['column']을 불러내면 안의 값이 나온다.
mask2 = ti['sibsp'] == 4              # 여기서 '=='을 사용했으므로 True 또는 False를 반환한다.
mask3 = ti['sibsp'] == 5

df_sib = ti[mask1 | mask2 | mask3]       # 조건식을 활용하여 데이터프레임을 만든다.
type(df_sib)
```

```python
df_sib.head()
```

```python
df_sib.info()
```

#  => isin() 메소드를 활용하여 변경

```python
# 위의 과정을 isin() 메소드를 활용하면 간단하게 끝낼 수 있다.

isin_filter = ti['sibsp'].isin([3, 4, 5])        # df['column']에서 원하는 조건을 isin()의 괄호 안에
df1_sib = ti[isin_filter]                        # 리스트(시리즈도 가능)의 형태로 넣어
df1_sib.head()                                   # 데이터프레임에 적용한다.
```

```python
df1_sib.count()              # count()를 사용하여 보면 39명이 조건에 부합하는 것을 확인할 수 있다.
```

# 컬럼만 추출이라는 조건이 붙은 경우는 loc를 사용

```python
# 승객의 나이가 가장 적거나, 많은 나이와 같은 승객만 추출, age, alive 컬럼만

age_col = ti['age'].isin([ti['age'].max(), ti['age'].min()])
df1_age = ti.loc[age_col, ['age', 'alive']]
df1_age
```

```python
# 승객의 나이가 20대에서 60대 사이, 함께 탑승한 형제가 3, 4, 5인 승객 중 
# 여성 승객의 나이, 성별, 형제 수, 요금을 출력

mask_age = (ti['age'] >= 20) & (ti['age'] <= 60)
mask_sib = ti['sibsp'].isin([3, 4, 5])
mask_female = ti['sex'] == 'female'

df_w = ti.loc[mask_age & mask_sib & mask_female, ['age', 'sex', 'sibsp', 'fare']]
df_w
```
