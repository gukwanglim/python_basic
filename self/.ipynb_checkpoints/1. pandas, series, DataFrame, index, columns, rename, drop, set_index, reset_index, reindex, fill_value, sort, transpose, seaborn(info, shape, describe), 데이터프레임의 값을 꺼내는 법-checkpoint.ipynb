# 판다스의 시리즈

```python
import pandas as pd           # 판다스 불러오기

# 시리즈 지정은 pd.Series({})
# ({}) 안에는 딕셔너리의 형태로 들어가며 key가 인덱스명, 값이 value로 적용

a_series = pd.Series({'a':10, 'b':20, 'c':30, 'd':40, 'e':30})

print(a_series); print()

print(type(a_series))                    # 타입은 시리즈로 나옴
```

# 시리즈의 데이터 선택 : 시리즈명[] -> 인덱스명 또는 인덱스 첨자로 찾을 수 있다

```python
print(a_series.index)                      # 시리즈의 index는 key에 속한다.
print()

print(a_series.values)                     # 시리즈의 value는 값에 속한다.
print()

print(a_series.value_counts())             # 딕셔너리 안의 같은 값의 갯수를 센다(데이터 분석용)
print()

print(a_series['a'], a_series[0])          # 인덱스는 간단하게 []으로 찾는다.
print()

print(a_series[[0, 1]])                    # 값이 여러개면 리스트로 넣는다. -> ['a', 'c'], [1, 2]
print()

print(a_series[0:2])                       # 숫자 인덱스 start:end -> end 미포함
print()

print(a_series['a':'c'])                   # 인덱스명 start:end -> end 포함
```

```python
b_serise = pd.Series([10, 'test', True, 'data'])
b_serise                      # 인덱스명을 주지 않으면 알아서 인덱스 포함
```

### 시리즈의 인덱스는 나중에 설정 가능

```python
c_serise = pd.Series([10, 'test', True, 'data'], index = ['a', 'b', 'c', 'd'])
c_serise                     # 첨자 인덱스로 인덱스 설정 가능
```

```python
d_serise = pd.Series([10, 'test', True, 'data'])
print(d_serise)
print()

d_serise.index = ['a', 'b', 'c', 'd']
print(d_serise)                                         # 이런 식으로도 인덱스를 넣을 수 있다.
```

# 데이터프레임 : 여러개의 시리즈가 모여서 만들어짐
### 딕셔너리의 키 = 칼럼명(시리즈 이름)이 됨
###   
### 주의 사항 
### 시리즈의 key = 인덱스명, 값 = value
### 데이터프레임의 key = 컬럼명, 값 = value    ->   인덱스명은 따로 정의

```python
# 데이터프레임 만드는 법 = pandas.DataFrame(딕셔너리, [index = 행이름], [colums = 열이름])

dict_data = {'c0': [1, 2, 3], 'c1' : [4, 5, 6], 'c2' : [7, 8, 9], 'c3' : [10, 11, 12]}

df = pd.DataFrame(dict_data)                     # 데이터프레임 생성
print(df); print()

print(df.index, df.columns); print()             # 인덱스, 컬럼 검색
                                                 # 아직 인덱스명을 정하지 않아 0, 1, 2로 나옴
df.index = ['a', 'b', 'c']               # 인덱스명 정의
print(df.index); print()

df.columns = ['a1', 'a2', 'a3', 'a4']    # 컬럼명 정의
print(df.columns)

df
```

# 행 인덱스 / 열 이름 지정

```python
df = pd.DataFrame([[15, '남', '덕영중'], [17, '여', '수리중']], 
                   index = ['준서', '예은'], columns = ['나이', '성별', '학교'])
print(df.index, df.columns)           # 처음 넣은 리스트는 데이터
                                      # 인덱스와 컬럼은 따로
```

### 행 이름 변경 : df.index = [인덱스명, ...]
### 컬럼명 변경 : df.columns = [컬럼명, ...]
### 행 인덱스의 일부 이름 변경 df.rename(index = {기존 인덱스명 : 새로운 인덱스명})
### 컬럼명의 일부 이름 변경 df.rename(cloumns = {기존 컬럼명 : 새로운 컬럼명})
###   
### 그냥 인덱스, 컬럼 뒤에는 [ ], () 안에 있는 인덱스, 컬럼 뒤에는 {}

```python
# 준서의 이름을 학생1로 변경

df.rename(index = {'준서' : '학생1'})     # 데이터를 바꿔 보여주지만 원본이 바뀐것은 아님
# print(df)                               # 준서가 인덱스로 들어가있음
```

```python
df.rename(index = {'준서' : '학생1'}, inplace = True)       
print(df); print()                         # inplace = True 기존 데이터 변경

df.rename(columns = {'학교' : '학교이름'}, inplace = True)
print(df)                                  # 컬럼 변경    
```

# 데이터프레임 삭제

```python
exam_data = {'수학' : [90, 80, 70], '영어' : [89, 98, 100],
            '음악' : [86, 97, 99], '체육' : [100, 90, 90]}

df = pd.DataFrame(exam_data)
print(df)
print()

# 데이터프레임 인덱스명 추가

df.index = ['서준', '우현', '인아']
print(df)

df2 = df.copy()           # df2 = df -> 주소만 복사되어 기존 데이터 공유
df2

# '우현' 자료 삭제 (행을 삭제)

df2.drop(['우현', '인아'], axis = 0, inplace = True)     # axis = 0 : 행(가로, 인덱스)을 삭제
                                        
    
df2.drop('수학', axis = 1)                              # axis = 1 : 열(세로, 컬럼)을 삭제
```

# 행 선택 -> df.loc[인덱스명], df.iloc[인덱스 첨자]

```python
exam_data = {'수학' : [90, 80, 70], '영어' : [89, 98, 100],
            '음악' : [86, 97, 99], '체육' : [100, 90, 90]}

df = pd.DataFrame(exam_data)

df2 = df.copy()

# df2의 인덱스는 '서준', '우현', '민아' 추가

df2.index = ['서준', '우현', '민아']
df2
```

```python
print(df2.loc['서준']); print()

df2.iloc[0]
```

```python
# '서준'과 '우현'의 자료 검색 : 인덱스 명으로

print(df2.loc[['서준', '우현']])          # 둘 이상을 찾을때는 리스트의 형태로
print()

print(df2.loc['서준' : '우현'])           # 우현 포함
print()

print(df2.iloc[0, 1])
print()

print(df2.iloc[0 : 1])                    # 1 미포함
print()

# 국어 점수를 100, 80, 90을 df2에 추가

print(df.수학)                            # 컬럼 자료 출력, .을 찍으면
print()                                   # []가 없어도 괜찮다.
                                          # 인덱스는 불가(loc, iloc 사용)
print(df['수학'])                                              
print()

df2['국어'] = [100, 80, 90]               # 컬럼 추가
df2
```

```python
# 새로운 행 추가 '철수' -> 100, 80, 90, 77, 50
print(df2)
df2.loc['철수'] = [100, 80, 90, 77, 50]          # 인덱스를 추가하기 때문에
df2                                              # loc 사용
```

### 각 원소 접근 df.loc[인덱스명, 컬럼명], df.loc[인덱스명][컬럼명]
### df.iloc[인덱스첨자, 컬럼위치 인덱스], df.iloc[인덱스첨자][컬럼위치]

```python
# 서준의 영어 점수 검색

print(df2.loc['서준', '영어'], df2.loc['서준']['영어'])  # 인덱스명이 포함되어
print()                                                  # loc, iloc가 필요.

print(df2.iloc[0, 1], df2.iloc[0][1])
print()

# 서준과 우현의 영어 점수

print(df2.loc[['서준', '우현'], '영어'])    # 두 개 이상일 경우 리스트
print()

# 서준의 영어 점수를 100점으로 시정
df2.loc['서준', '영어'] = 100
df2
print()

# 서준부터 민아까지의 수학과 영어 점수를 출력

print(df2.loc[['서준', '우현', '민아'], ['수학', '영어']])
df2.loc['서준' : '민아', ['수학', '영어']]         # 인덱스는 ':' 사용 가능
                                                   # 컬럼은 ':' 사용 불가
                                       # but, iloc 일 경우 둘 다 ':' 사용가능
                                       # ex) df2.iloc[0 : 2, 1 : 3]
```

# 데이터프레임 : 2차원 배열 (시리즈가 열로 묶인 배열)
### 인덱스 접근 방법 : df.loc[인덱스명], 값이 여러개 [ , ...]
### 컬럼 접근 방법 : df[컬럼명]
### 원소 접근 방법 : df.loc[인덱스명, 컬럼명]
###                            df.iloc[인덱스 첨자, 컬럼 첨자]
### 인덱스 생성 : df.index = [인덱스명, ...]
### 특정 칼럼을 인덱스로 : df.set_index(컬럼명, inplace = True)
### 인덱스 초기화 : df.reset_index(inplace = True)
### 인덱스 재배열 : df.reindex([인덱스명, ...])
### 인덱스 삭제 : df.drop(index = [인덱스명])
### 컬럼 삭제 : df.drop(columns = [컬럼명])
### 인덱스 정렬 : df.sort_index()
### 값으로 정렬 : df.sort_values(by = 컬럼명, ascending = False)

```python
print(df2); print()

print(df2.iloc[0 : 2, 1 : 3])                  # 0행, 1행, 1열 2열
print()

print(df2.iloc[:2, :]); print()

print(df2.iloc[: , :])               # ':'을 사용할 때에는 리스트의 형태가 아님
```

# 시리즈 연산 : 시리즈 + 사칙연산 + 숫자

```python
sr = pd.Series({'국어' : 100, '수학' : 90, '영어' : 90})
print(sr); print()

# 각 점수에 100을 더한 값 출력

print(sr+ 100); print()

# 각 점수에 100을 나눈 값 출력

print(sr / 100)
```

# 시리즈 + 사칙연산 + 시리즈
### NaN 처리 필수!

```python
sr1 = pd.Series({'국어' : 100, '수학' : 90, '영어' : 90, '체육' : 80})
sr2 = pd.Series({'영어' : 100, '국어' : 90, '수학' : 90, '과학' : 90})      # sr1에 과학이 없으므로
                                                                            # 과학은 NaN이 나온다.
print(sr1 + sr2); print()

print(sr1.add(sr2, fill_value = 0))                   # sr1에다가 sr2의 값을 더해라. 대신 nan 값이 
                                                      # 있다면 nan 대신 0을 넣어라.
```

```python
# 문) titanic 데이터셋을 불러와
# 1. 나이가 10세 미만인 승객의 class, 성별, 나이, 생존여부를 출력
# 2. 클래스가 1등급인 승객의 나이, fare, 클래스, 생존여부 출력
# 3. 성별이 남성이고 20대에서 40대 미만인 승객의 모든 정보 출력
```

```python
df_ti = sns.load_dataset('titanic')
df_ti.info
```

```python
df_ti.columns
```

```python
df_ti_1 = df_ti.loc[df_ti.age < 10, ['class', 'sex', 'age', 'alive']]     
df_ti_1           # age는 컬럼이지만 인덱스의 위치로 들어가게 자료 출력
         # 원래라면 loc 다음에 인덱스가 들어와야 하지만, 
        # 저장된 자료에 뽑아내고 싶은 조건이 있을 경우 이런 방식으로 작성
```

### 컬럼을 뽑아내는 두 가지 방법:
### 1. df.수학
### 2. df['수학']

```python
df_ti_2 = df_ti.loc[df_ti['class'] == 'First', ['age', 'fare', 'class', 'alive']]
df_ti_2         # class는 명령어가 있으므로 df.class의 형식으로 사용 불가
```

```python
df_ti_3 = df_ti.loc[(df_ti.sex == 'male') & (df_ti.age >= 20) & (df_ti.age < 40), :]
df_ti_3     # 그리고(and)를 사용하기 위해서는 '&' 을 사용한다.
            # 또한, () & ()의 형식으로 사용해야한다.
```
