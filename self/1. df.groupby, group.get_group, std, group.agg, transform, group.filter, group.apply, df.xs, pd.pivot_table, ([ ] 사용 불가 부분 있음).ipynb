# 1. 그룹 만들기 : df.groupby(기준이 되는 컬럼명) : 그룹이 여러개면 [리스트,...]

```python
# 선실의 레벨별(class)로 승객수를 산출

grouped = ti.groupby('class')           # titanic 데이터프레임에서 'class'를 기준으로 그룹을 만든다.
grouped
```

```python
for key, group in grouped:
    print('key : ', key)            # 그룹의 키는 내가 지정한 그룹을 이루는 값(First, Second, Third)
    print('승객 수 : ', len(group))      # group은 해당 키에 속하는 인원수
    print(grouped.head())           # 해당 키에 속하는 리스트의 위 5개
```

# group.agg(매핑함수)

```python
# 집계연산을 처리 -> 사용자 함수를 그룹 객체에 적용

ti2 = ti.loc[ : , ['survived', 'age', 'class', 'fare']]

ti2_g = ti2.groupby('class')

def max_min(x):
    return x.max() - x.min()
```

```python
# group에 max_min()함수를 적용

ti2_g_agg = ti2_g.agg(max_min)
ti2_g_agg.head()
```

# 모든 열에 여러 함수를 매핑 : 그룹객체.agg(['함수1', '함수2', ...]) 
                                                           -> 함수를 여러개 사용할 때는 리스트
# 각 열마다 다른 함수를 매핑 : 그룹객체.agg({'열1' : '함수1', '열2' : '함수2', ...})
                                                           -> 여기서는 딕셔너리의 형태로

```python
agg_all = ti2_g.agg(['max', 'min'])
agg_all
```

```python
agg_sep = ti2_g.agg({'age' : 'mean', 'fare' : ['min', 'max']})
agg_sep                                      # 다른 컬럼에 여러개의 함수를 넣을 때는 리스트
```

# 그룹 객체.transform(매핑함수) : 각 그룹의 행에 대한 연산

```python
age_mean = ti2_g.age.mean()       # 그룹별 'age' 컬럼의 평균
age_mean.head()                                             # 그룹의 key가 인덱스로 설정됨
```

```python
age_mean.loc['First']
```

```python
age_std = ti2_g.age.std()
age_std
```

```python
# 그룹 객체의 age 컬럼에 대해 zscore 계산하여 출력

for key, group in ti2_g.age:
    group_zscore = (group - age_mean.loc[key]) / age_std.loc[key]
    print('origin : ', key)
    print(group_zscore.head(3))
```

```python
def z_score(x):
    return (x - x.mean()) / x.std()

age_zscore = ti2_g.age.transform(z_score)
age_zscore.head()
```

```python
age_zscore.loc[0 : 9]
```

# 그룹 객체 필터링 : 그룹객체.filter(조건식 함수)

```python
# class로 그룹을 지은 후, 그룹별 데이터의 개수가 200이상인 그룹만 추출

ti2_g_filter = ti2_g.filter(lambda x: len(x) >= 200)
ti2_g_filter
```

```python
ti2_g_filter['class'].unique()
```

```python
# age 컬럼의 평균값이 30세보다 적은 그룹만 추출

age_filter = ti2_g.filter(lambda x: x.age.mean() < 30)
age_filter
```

```python
age_filter['class'].unique()
```

# xs() 사용하기

```python
# values : 'age', 'fare'
# aggfunc : 'mean', 'sum'
```

```python
pdf2 = pd.pivot_table(df, index = ['class', 'sex'],   # 행 위치에 들어갈 컬럼
                     columns = 'survived',         # 열 위치에 들어갈 컬럼
                     values = ['age', 'fare'],            # 데이터로 사용할 컬럼
                     aggfunc = ['mean', 'sum']
                     )
pdf2
```

```python
pdf2.xs(('Second', 'male'), level = [0, 'sex'])    # level = [인덱스, 컬럼명]
```

```python
pdf2.xs(('Second', 'male'))                # level을 설정하지 않으면 위처럼 표의 형태가 아닌 
                                           # 문장의 형태로 출력된다.
```

```python
pdf2.xs(('mean', 'age'), axis = 1)
```

```python
pdf2.xs('mean', axis = 1)
```
